## **III. Data init**

````python
# Danh sách sinh viên ban đầu
students = [
    {
        "id": "SV001",
        "name": "Nguyễn Văn An",
        "age": 20,
        "scores": {"SAD": 8.5, "DSA": 7.0, "BigData": 9.0}
    },
    {
        "id": "SV002",
        "name": "Trần Thị Bình",
        "age": 21,
        "scores": {"SAD": 6.5, "DSA": 7.5, "BigData": 6.0}
    },
    {
        "id": "SV003",
        "name": "Lê Văn Cường",
        "age": 19,
        "scores": {"SAD": 9.0, "DSA": 8.5, "BigData": 9.5}
    },
    {
        "id": "SV004",
        "name": "Phạm Thị Dung",
        "age": 20,
        "scores": {"SAD": 5.0, "DSA": 4.5, "BigData": 6.0}
    },
    {
        "id": "SV005",
        "name": "Hoàng Văn Em",
        "age": 22,
        "scores": {"SAD": 7.0, "DSA": 7.0, "BigData": 7.5}
    }
]


print(f"Số lượng sinh viên: {len(students)}")
````

````python
def calculate_average(scores):
    """
    Tính điểm trung bình từ dictionary điểm
    
    Args:
        scores (dict): Dictionary chứa điểm các môn
    
    Returns:
        float: Điểm trung bình
    """
    # code của bạn
    avg = sum(scores.values()) / len(scores)
    return avg

test_scores = {
    "SAD": 7.0,
    "DSA": 7.0,
    "BigData": 7.5
}

print(f"Điểm trung bình: {calculate_average(test_scores):.2f}")
````

## **IV. CÁC CHỨC NĂNG**

### 1: Hiển thị danh sách sinh viên

````python
def display_all_students():
    """
    Hiển thị danh sách tất cả sinh viên
    In ra: ID, Tên, Tuổi, Điểm trung bình
    """
    # code của bạn
    print("---DANH SÁCH SINH VIÊN---\n")
    for i in students:
        print(f"Mã sinh viên: {i['id']}")
        print(f"Tên sinh viên: {i['name']}")
        print(f"Tuổi sinh viên: {i['age']}")
        print(f"Điểm: {i['scores']}")
        print("\n------\n")

# Test 
display_all_students()
````

### 2: Thêm sinh viên mới

````python
def add_new_student():
    """
    Thêm sinh viên mới
    - Tự động tạo ID theo format SV001, SV002...
    - Nhập: name, age
    - Nhập điểm 3 môn: SAD, DSA, BigData
    """
    
    # code của bạn
    
    id = f"SV{len(students) + 1:03d}"   
    name = input("Nhập tên sinh viên: ")
    age = int(input("Nhập tuổi: "))

    sad = float(input("Nhập điểm SAD: "))
    dsa = float(input("Nhập điểm DSA: "))
    bigData = float(input("Nhập điểm BigData: "))

    new_student = {
        "id": id,
        "name": name,
        "age": age,
        "scores": {"SAD": sad, "DSA": dsa, "BigData": bigData}
    }
    
    students.append(new_student)

add_new_student()
````

### 3: Cập nhật điểm

````python
def update_scores():
    """
    Cập nhật điểm sinh viên
    - Nhập ID sinh viên
    - Hiển thị điểm hiện tại
    - Cho phép cập nhật từng môn (Enter để bỏ qua)
    - Cập nhật dictionary scores
    """
    print("\n--- CẬP NHẬT ĐIỂM ---")
    
    # code của bạn
    up_id = input("Nhập id cần cập nhật điểm: ")
    found = False

    for i in students:
        if i["id"] == up_id:
            
            print("Điểm hiện tại: ", i["scores"])

            #Cập nhật từng môn
            for p in i["scores"]:
                new_score = input("Nhập điểm mới lần lượt cho 3 môn: ")

            found = True
            print("Cập nhật thành công")
            #print("Điểm sau khi cập nhật: ", i["scores"])
            break

    if not found:
        print(f"Không tìm thấy sinh viên có {up_id} để cập nhật")

update_scores()
````

### 4: Xóa sinh viên theo ID

````python
def delete_student():
    """
    Xóa sinh viên theo ID 
    """
    print("\n--- XÓA SINH VIÊN ---")

    # code của bạn
    re_id = input("Nhập id cần xóa: ")
    found = False

    for i in students:
        if i["id"] == re_id:
            students.remove(i)
            print(f"Xóa thành công sinh viên có id:  {i['id']}")
            found = True
            break

    if not found:
        print(f"Không có {re_id} để xóa!")

delete_student()
````

### 5: Tìm sinh viên theo tên

````python
def search_by_name():
    """
    Tìm sinh viên theo tên
    - Không phân biệt hoa thường
    - Tìm kiếm gần đúng (chứa từ khóa)
    """
    print("\n--- TÌM SINH VIÊN THEO TÊN ---")
    
    find_name = input("Nhập tên sinh viên cần tìm: ")
    found = False

    for i in students:
        if find_name.lower() in i["name"].lower():
            print(f"Đã tìm thấy sinh viên {i['id']} - {i['name']}")
            found = True
            break

    if not found:
        print("Không tìm thấy sinh viên: ", find_name)

search_by_name()
````

### 6: Xếp hạng sinh viên

````python
def rank_students():
    """
    Xếp hạng sinh viên theo điểm trung bình
    - Tính điểm TB cho tất cả sinh viên
    - Sắp xếp theo điểm TB giảm dần
    - Sử dụng sorted() với key=lambda
    - Hiển thị bảng xếp hạng có số thứ tự
    """
    print("\n--- BẢNG XẾP HẠNG SINH VIÊN ---")
    
    # code của bạn

    # Tính điểm trung bình cho từng sinh viên
    ranked = []
    for i in students:
        scores = i["scores"].values()
        avg = sum(scores) / len(scores)
        ranked.append({
            "id": i["id"],
            "name": i["name"],
            "avg": avg
        })
    
    # Sắp xếp theo điểm TB giảm dần
    # Dùng sorted() để sắp xếp dsach ranked, key=lambda x: x["avg"] sắp xếp theo gtri avg
    # reverse = True/False sắp xếp giảm/tăng dần
    ranked = sorted(ranked, key=lambda x: x["avg"], reverse = True)
    
    # Hiển thị bảng xếp hạng
    print(f"{'STT':<5} {'ID':<10} {'Tên':<20} {'Điểm TB':<10}")
    for idx, s in enumerate(ranked, start = 1):
        print(f"{idx:<5} {s['id']:<10} {s['name']:<20} {s['avg']:<10.2f}")

rank_students()
````

### 7: Thống kê theo mức điểm

````python
def statistics_by_grade():
    """
    Thống kê sinh viên theo mức điểm
    - Giỏi: điểm TB >= 8.0
    - Khá: 6.5 <= điểm TB < 8.0
    - Trung bình: 5.0 <= điểm TB < 6.5
    - Yếu: điểm TB < 5.0
    - Đếm số sinh viên mỗi loại
    """
    print("\n--- THỐNG KÊ THEO MỨC ĐIỂM ---")
    
    # code của bạn
    gioi = kha = tb = yeu = 0

    for i in students:
        scores = i["scores"].values()
        avg = sum(scores) / len(scores)

        if avg >= 8.0:
            gioi += 1
        elif (avg >= 6.5) and (avg < 8.0):
            kha += 1
        elif (avg >= 5.0) and (avg < 6.5):
            tb += 1
        else:
            yeu += 1

    print(f"Sinh viên giỏi       : {gioi}")
    print(f"Sinh viên khá        : {kha}")
    print(f"Sinh viên trung bình : {tb}")
    print(f"Sinh viên yếu        : {yeu}")

statistics_by_grade()
````

### 8: Tìm sinh viên điểm cao nhất/thấp nhất

````python
def find_top_bottom_students():
    """
    Tìm sinh viên có điểm cao nhất và thấp nhất
    - Sử dụng max()/min() với key
    - Hiển thị thông tin chi tiết
    """
    print("\n--- SINH VIÊN ĐIỂM CAO NHẤT VÀ THẤP NHẤT ---\n")

    # code của bạn
    # Hàm tính điểm trung bình
    def DiemTB(scores):
        return sum(scores.values()) / len(scores)

    # Tìm sinh viên điểm cao nhất
    caoI = max(students, key=lambda s: DiemTB(s["scores"]))
    # Tìm sinh viên điểm thấp nhất
    thapI = min(students, key=lambda s: DiemTB(s["scores"]))

    # Hiển thị kết quả
    print("Sinh viên điểm cao nhất:")
    print(f"ID: {caoI['id']}, Tên: {caoI['name']}, "
          f"Điểm TB: {DiemTB(caoI['scores']):.2f}\n")

    print("Sinh viên điểm thấp nhất:")
    print(f"ID: {thapI['id']}, Tên: {thapI['name']}, "
          f"Điểm TB: {DiemTB(thapI['scores']):.2f}")

find_top_bottom_students()
````

````python
def main_menu():
    while True:
        print("\n" + "="*50)
        print("     CHƯƠNG TRÌNH QUẢN LÝ SINH VIÊN VÀ ĐIỂM")
        print("="*50)
        print("1. Hiển thị danh sách sinh viên")
        print("2. Thêm sinh viên mới")
        print("3. Cập nhật điểm")
        print("4. Xóa sinh viên theo ID")
        print("5. Tìm sinh viên theo tên")
        print("6. Xếp hạng sinh viên")
        print("7. Thống kê theo mức điểm")
        print("8. Tìm sinh viên điểm cao nhất/thấp nhất")
        print("0. Thoát")
        print("="*50)
        
        choice = input("Chọn chức năng (0-8): ").strip()
        
        if choice == "1":
           # code của bạn 
           display_all_students()
        elif choice == "2":
            add_new_student()          
        elif choice == "3":
            update_scores()              
        elif choice == "4":
            delete_student()      
        elif choice == "5":
            search_by_name()       
        elif choice == "6":
            rank_students()              
        elif choice == "7":
            statistics_by_grade()        
        elif choice == "8":
            find_top_bottom_students()   
        elif choice == "0":
            print("\n✓ Thoát chương trình. Tạm biệt!")
            break

        else:
            print("\n✗ Lựa chọn không hợp lệ! Vui lòng chọn lại.")

main_menu()
````
