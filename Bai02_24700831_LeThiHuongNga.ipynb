## II. CẤU TRÚC DỮ LIỆU

Mỗi cuốn sách được lưu trữ dưới dạng Dictionary với cấu trúc:

```python
{
    "id": 1,
    "title": "Tên sách",
    "author": "Tác giả",
    "year": 2024,
    "status": "available"  # available hoặc borrowed
}
```

````python
books = [
    {"id": 1, "title": "Lập trình Python cơ bản", "author": "Nguyễn Văn A", "year": 2020, "status": "available"},
    {"id": 2, "title": "Khoa học dữ liệu với Python", "author": "Trần Thị B", "year": 2021, "status": "borrowed"},
    {"id": 3, "title": "Machine Learning cơ bản", "author": "Lê Văn C", "year": 2022, "status": "available"},
    {"id": 4, "title": "Python nâng cao", "author": "Nguyễn Văn A", "year": 2023, "status": "available"},
    {"id": 5, "title": "Big Data với Python", "author": "Phạm Thị D", "year": 2024, "status": "borrowed"}
]

print("Đã khởi tạo dữ liệu mẫu thành công!")
print(f"Số lượng sách: {len(books)}")
````

````python
def display_all_books():
    """
    Hiển thị danh sách tất cả các sách trong thư viện
    In ra: ID, Tên sách, Tác giả, Năm xuất bản, Trạng thái
    """
    # code của bạn
    print("---DANH SÁCH TẤT CẢ CÁC SÁCH---\n")
    for i in books:
        print(f"Mã sách: {i['id']}")
        print(f"Mã sách: {i['title']}")
        print(f"Mã sách: {i['author']}")
        print(f"Mã sách: {i['year']}")
        print(f"Mã sách: {i['status']}")
        print("\n------\n")

display_all_books()
````

````python
def add_new_book():
    """
    Thêm sách mới vào thư viện
    - Tự động tạo ID
    - lấy giá trị truyền vô: title, author, year
    - Mặc định status = "available"
    """
    print("\n--- THÊM SÁCH MỚI ---")
    
    # code của bạn
    new_id = books[-1]["id"] + 1 if books else 1  
    title = input("Nhập tên sách: ")
    author = input("Nhập tên tác giả: ")
    year = int(input("Nhập năm xuất bản: "))
    status = "available"

    new_book = {
        "id": new_id,
        "title": title,
        "author": author,
        "year": year,
        "status": status
    }
    
    books.append(new_book)
    print(f"Đã thêm sách '{title}' (ID: {new_id}) thành công!")

add_new_book()
````

````python
def borrow_book():
    """
    Mượn sách từ thư viện
    - Nhập ID sách cần mượn
    - Kiểm tra status = "available"
    - Cập nhật status = "borrowed"
    - Thông báo kết quả
    """
    print("\n--- MƯỢN SÁCH ---")
    
    # code của bạn
    find_id = int(input("Nhập ID sách cần mượn: "))
    found = False

    for b in books:
        if b['id'] == find_id:
            found = True
            if b['status'] == "available":
                b['status'] = "borrowed"
                print(f"Đã mượn sách có id {b['id']} - tên {b['title']} thành công!")
            else:
                print(f"Sách có id {b['id']} -tên {b['title']} đã bị mượn.")
            break

    if not found:
        print(f"Không tìm thấy sách với ID: {find_id}")

borrow_book()
````

````python
def return_book():
    """
    Trả sách cho thư viện
    - Nhập ID sách cần trả
    - Kiểm tra status = "borrowed"
    - Cập nhật status = "available"
    - Thông báo kết quả
    """
    print("\n--- TRẢ SÁCH ---")
    
    # code của bạn
    find_id = int(input("Nhập ID sách cần trả: "))
    found = False

    for b in books:
        if b["id"] == find_id:
            found = True
            if b["status"] == "borrowed":
                b["status"] = "available"
                print(f"Đã trả sách thành công")
            else:
                print(f"Sách tên '{b['title']}' chưa bị mượn")
            break

    if not found:
        print(f"Không tìm thấy sách có ID: {find_id}")

return_book()
````

````python
def search_by_author():
    """
    Tìm kiếm sách theo tác giả
    - Nhập tên tác giả
    - Duyệt list và hiển thị tất cả sách của tác giả đó
    - Sử dụng lower() để không phân biệt hoa thường
    """
    print("\n--- TÌM SÁCH THEO TÁC GIẢ ---")
    
    # code của bạn 

    find_au = input("Nhập tên tác giả cần tìm: ")
    found = False

    for b in books:
    
        if find_au.lower() in b["author"].lower():
            print(f"ID: {b['id']} | Tên sách: {b['title']} | Tên tác giả: {b['author']}")
            found = True

    if not found:
        print(f"Không tìm thấy sách của tác giả {find_au}")

search_by_author()
````

````python
def filter_by_year():
    """
    Lọc sách theo khoảng năm xuất bản
    - Nhập năm bắt đầu và năm kết thúc
    - Hiển thị các sách trong khoảng năm đó
    """
    print("\n--- LỌC SÁCH THEO NĂM XUẤT BẢN ---")
    
    # code của bạn
    start_year = int(input("Nhập năm bắt đầu: "))
    end_year = int(input("Nhập năm kết thúc: "))
    found = False

    for b in books:
        if (start_year <= b["year"]) and (b["year"] <= end_year):
            print(f"ID: {b['id']} | Tên sách: {b['title']} | Năm: {b['year']}")
            found = True

    if not found:
        print(f"Không tìm thấy sách trong khoảng {start_year} - {end_year}")

filter_by_year()
````

````python
def show_statistics():
    """
    Hiển thị thống kê thư viện
    - Tổng số sách
    - Số sách đang available
    - Số sách đang borrowed
    - Tính phần trăm
    """
    print("\n--- THỐNG KÊ THƯ VIỆN ---")
    
    # code của bạn 
    # Tổng số sách
    s = len(books)

    # số sách theo trạng thái available/ borrowed
    ava, bor = 0, 0

    for b in books:
        if b["status"] == "available":
            ava += 1
        elif b["status"] == "borrowed":
            bor += 1

    # Tính phần trăm 
    if s > 0:
        avai_per = (ava / s) * 100
        bor_per = (bor / s) * 100
    else:
        avai_per = bor_per= 0

    print(f"Tổng số sách: {s}")
    print(f"Sách đang available: {ava} ({avai_per:.2f}%)")
    print(f"Sách đang borrowed: {bor} ({bor_per:.2f}%)")

show_statistics()
````

````python
def main_menu():
    # code của bạn
    while True:
        print("\n" + "="*50)
        print("     CHƯƠNG TRÌNH QUẢN LÝ SÁCH      ")
        print("="*50)
        print("1. Hiển thị danh sách tất cả sách")
        print("2. Thêm sách mới")
        print("3. Mượn sách theo id")
        print("4. Trả sách theo id")
        print("5. Tìm sách theo tên tác giả")
        print("6. Lọc sách theo năm xuất bản")
        print("7. Thống kê thư viện")
        print("0. Thoát")
        print("="*50)

        choice = input("Chọn chức năng (0-7): ").strip()
        
        if choice == "1":
           display_all_books()
        elif choice == "2":
            add_new_book()          
        elif choice == "3":
            borrow_book()              
        elif choice == "4":
            return_book()     
        elif choice == "5":
            search_by_author()       
        elif choice == "6":
            filter_by_year()              
        elif choice == "7":
            show_statistics()        
        elif choice == "0":
            print("\nThoát chương trình. Tạm biệt!")
            break

        else:
            print("\nLựa chọn không hợp lệ! Vui lòng chọn lại.")
````

````python
main_menu()
````
