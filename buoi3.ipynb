```python
class sinhvien:
    def __init__(self, maSV,hoTen, diaChi,diemA,diemB):
        self.maSV=maSV 
        self.hoTen=hoTen 
        self.diaChi=diaChi 
        self.diemA=diemA 
        self.diemB=diemB 
    def tinhTB(self):
        return(self.diemA+self.diemB)/2
        
```

```python
class SinhvienKSTN (sinhvien):
    def __init__(self, maSV, hoTen, diaChi, diemA, diemB, diemC):
        super().__init__(maSV, hoTen, diaChi, diemA, diemB)
        self.diemC=diemC
    def tinhTB(self):
        return (self.diemA+ self.diemB+self.diemC)/3
```
