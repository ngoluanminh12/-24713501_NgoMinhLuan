# BÀI TẬP CÁC KIỂU DỮ LIỆU

**1. Kiểu dữ liệu cơ bản:**
- int, float: Các phép toán số học
- str: Xử lý chuỗi, các phương thức string
- bool: Logic và điều kiện

**2. Collection types:**
- **List**: Có thứ tự, có thể thay đổi, cho phép trùng lặp → Dùng khi cần thay đổi dữ liệu
- **Tuple**: Có thứ tự, KHÔNG thể thay đổi, cho phép trùng lặp → Dùng cho dữ liệu cố định
- **Set**: KHÔNG có thứ tự, có thể thay đổi, KHÔNG cho phép trùng lặp → Dùng để loại bỏ duplicate
- **Dictionary**: Lưu trữ cặp key-value, có thể thay đổi → Dùng khi cần mapping

**3. Khi nào dùng gì:**
- Cần lưu danh sách có thể thay đổi → **List**
- Cần lưu dữ liệu không đổi (như tọa độ) → **Tuple**
- Cần loại bỏ phần tử trùng → **Set**
- Cần mapping (ví dụ: tên → điểm) → **Dictionary**

**Bài 3.9:** Mã hóa Caesar Cipher (dịch mỗi ký tự đi k vị trí)

Ví dụ: "abc", k=2 → "cde"

````python
# Code 
def caesar_cipher(text, k):
    result = ""
    for char in text:
        if char.isalpha():  # chỉ mã hóa chữ cái
            # Xác định bảng chữ cái
            base = ord('A') if char.isupper() else ord('a')
            # Dịch chuyển k vị trí
            result += chr((ord(char) - base + k) % 26 + base)
        else:
            # Giữ nguyên ký tự không phải chữ cái
            result += char
    return result

# Ví dụ
print(caesar_cipher("abc", 2))       # cde
print(caesar_cipher("xyz", 3))       # abc
````
