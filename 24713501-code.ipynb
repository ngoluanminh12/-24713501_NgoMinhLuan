```python
def themsp():
    n=int(input("nhap so sp: "))
    for i in range(1, n+1):
        ten = input("nhap ten sp: ")
        gia= float(input("nhap gia: "))
        mota= input("nhap mo ta: ")
        sp={
            "id": id,
            "name": ten,
            "price": gia,
            "description":mota
        }
        product.append(sp)
```

```python

def xoatheoid():
    find_id= int(input("nhap id muon xoa: "))
    found = False
    for p in product:
        if p['id']==find_id:
            product.remove(p)
            print("da xoa thanh cong")
            found= True
            break
    if not found:
            print("khong tim thay id")
```

```python
def capnhat():
    id= int(input("nhap id "))

    for p in product:
        if p['id'] == id:
            name=input("nhap ten moi")
            price=float(input("nhap gia"))
            description=input("nhap mo ta")
            if name:
                p['name'] = name
            if price:
                p['price']= price
            if description:
                p['description'] = description
            print("cap nhat thanh cong")
            break
        else:
            print("khong tim thay id")
            break
        
capnhat()
        

```
